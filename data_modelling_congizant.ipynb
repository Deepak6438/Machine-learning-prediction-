Now, Apply the machine learning model , Apply linear regression model, because the target variable is continues variables.

```python
# define target (y) and features (x)
```

```python
x = analysis_data.drop(columns="estimated_stock_pct")
y = analysis_data['estimated_stock_pct']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split (x,y, train_size=0.8,random_state=2529)
```

```python
# check shape size
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# import machine learning model
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train and fit the model
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# predict the model
```

```python
model.predict(x_test)
```

```python
y_pred = model.predict(x_test)
```

```python
# import model accuracy
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
